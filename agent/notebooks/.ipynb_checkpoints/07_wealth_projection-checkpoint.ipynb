# 07 — Wealth Growth & 10-Year Net Worth Projection

This notebook tracks your portfolio's growth over time and projects net worth toward financial independence.

- **Growth chart** — month-over-month portfolio value from historical snapshots
- **Growth metrics** — MoM return, annualized return (CAGR)
- **10-year projection** — compound growth with monthly contributions and income growth
- **Milestone tracker** — when you'll hit $250K, $500K, $1M

**Prerequisites:**

- `portfolio_snapshot.json` must exist for current portfolio value

## Your Financial Profile

Adjust these variables to match your situation. Re-run the notebook after changes.

**Brokerage portfolio** is tracked automatically from Plaid data.
**Other assets** (retirement, cash) are entered manually below — update them periodically.

```python
# === YOUR FINANCIAL PROFILE ===
CURRENT_AGE = 31
TARGET_FI_AGE = 45                       # Financial independence target age
MONTHLY_CONTRIBUTION = 1000              # $ invested per month into brokerage accounts
ANNUAL_CONTRIBUTION_INCREASE = 5.0       # % yearly increase in contributions (income growth)
ASSUMED_ANNUAL_RETURN = 8.0              # % — used when not enough history to calculate

# === OTHER ASSETS (update periodically) ===
# Retirement + HSA not tracked by Plaid (total investments $257,959 minus Plaid-tracked brokerage)
RETIREMENT_HSA = 85_711                  # 401k, IRA, HSA — untracked portion
RETIREMENT_MONTHLY_CONTRIBUTION = 0      # $ monthly into retirement/HSA (employer + you)
RETIREMENT_ANNUAL_RETURN = 7.0           # % assumed return for retirement/HSA accounts
CASH_SAVINGS = 29_449                    # Checking, savings, emergency fund
DEBTS = 7_086                            # Credit card balance
# ==========================================

# Verified net worth: $280,322 = $29,449 cash - $7,086 debt + $257,959 investments
# Plaid tracks $172,248 of the $257,959 in investments automatically
# The remaining $85,711 (retirement/HSA) is entered manually above

PROJECTION_YEARS = TARGET_FI_AGE - CURRENT_AGE
print(f"Profile: age {CURRENT_AGE}, targeting FI at {TARGET_FI_AGE} ({PROJECTION_YEARS} years)")
print(f"Monthly brokerage contribution: ${MONTHLY_CONTRIBUTION:,} (growing {ANNUAL_CONTRIBUTION_INCREASE}%/yr)")
print(f"Assumed brokerage return: {ASSUMED_ANNUAL_RETURN}% annual")
print(f"Retirement/HSA (untracked): ${RETIREMENT_HSA:,} (growing at {RETIREMENT_ANNUAL_RETURN}%/yr)")
print(f"Cash: ${CASH_SAVINGS:,}  |  Debts: ${DEBTS:,}")
```

```python
import json
import os
from datetime import datetime, timedelta

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as mticker

# Dark theme to match project aesthetic
plt.style.use("dark_background")
plt.rcParams.update({
    "figure.facecolor": "#1a1a2e",
    "axes.facecolor": "#16213e",
    "axes.edgecolor": "#333",
    "axes.labelcolor": "#ccc",
    "xtick.color": "#888",
    "ytick.color": "#888",
    "grid.color": "#2a2a4a",
    "text.color": "#e0e0e0",
    "figure.figsize": (12, 6),
})

# Load portfolio history
HISTORY_FILE = os.path.join("..", "portfolio_history.json")
SNAPSHOT_FILE = os.path.join("..", "portfolio_snapshot.json")

if os.path.exists(HISTORY_FILE):
    with open(HISTORY_FILE, "r") as f:
        history = json.load(f)
else:
    history = []

if not os.path.exists(SNAPSHOT_FILE):
    raise FileNotFoundError("portfolio_snapshot.json not found. Run notebook 02 first.")

with open(SNAPSHOT_FILE, "r") as f:
    snapshot = json.load(f)

current_value = snapshot["summary"]["total_value"]
current_cost = snapshot["summary"]["total_cost_basis"]

print(f"\u2705 Current portfolio: ${current_value:,.2f}")
print(f"\u2705 History entries: {len(history)}")

if history:
    hist_df = pd.DataFrame(history)
    hist_df["date"] = pd.to_datetime(hist_df["date"])
    hist_df = hist_df.sort_values("date").reset_index(drop=True)
    print(f"   Date range: {hist_df['date'].iloc[0].strftime('%Y-%m-%d')} to {hist_df['date'].iloc[-1].strftime('%Y-%m-%d')}")
```

```python
# === NET WORTH SNAPSHOT ===
total_investments = current_value + RETIREMENT_HSA
total_net_worth = CASH_SAVINGS + total_investments - DEBTS

print("=" * 60)
print("         CURRENT NET WORTH")
print("=" * 60)
print(f"  Investments:           ${total_investments:>12,.2f}")
print(f"    Brokerage (Plaid):     ${current_value:>12,.2f}")
print(f"    Retirement/HSA:        ${RETIREMENT_HSA:>12,.2f}")
print(f"  Cash / savings:        ${CASH_SAVINGS:>12,.2f}")
if DEBTS > 0:
    print(f"  Credit card debt:     -${DEBTS:>12,.2f}")
print(f"  {'─' * 42}")
print(f"  TOTAL NET WORTH:       ${total_net_worth:>12,.2f}")
print("=" * 60)
debt_status = "Debt-free" if DEBTS == 0 else f"${DEBTS:,.0f} CC debt"
print(f"\n  Age {CURRENT_AGE} | {debt_status} | {PROJECTION_YEARS} years to FI target (age {TARGET_FI_AGE})")
```

## Growth Metrics

```python
effective_annual_return = ASSUMED_ANNUAL_RETURN  # default

print("=" * 60)
print("         GROWTH METRICS")
print("=" * 60)

if history and len(hist_df) >= 2:
    first = hist_df.iloc[0]
    last = hist_df.iloc[-1]
    days_elapsed = (last["date"] - first["date"]).days

    total_return_pct = (last["total_value"] / first["total_value"] - 1) * 100
    print(f"  Total return over period:  {total_return_pct:+.2f}%")
    print(f"  Period: {days_elapsed} days ({first['date'].strftime('%Y-%m-%d')} to {last['date'].strftime('%Y-%m-%d')})")

    if days_elapsed >= 90:  # 3+ months — use CAGR
        years_elapsed = days_elapsed / 365.25
        cagr = ((last["total_value"] / first["total_value"]) ** (1 / years_elapsed) - 1) * 100
        effective_annual_return = max(cagr, 0)  # floor at 0%
        print(f"  Annualized return (CAGR): {cagr:+.2f}%")
        print(f"\n  Using calculated CAGR ({effective_annual_return:.1f}%) for projection.")
    else:
        print(f"\n  Not enough history for reliable CAGR (need 3+ months, have {days_elapsed} days).")
        print(f"  Using assumed return ({ASSUMED_ANNUAL_RETURN}%) for projection.")
else:
    print(f"  Not enough data points for growth metrics (have {len(history)}, need 2+).")
    print(f"  Using assumed return ({ASSUMED_ANNUAL_RETURN}%) for projection.")
    print(f"\n  Run the pipeline again next month to start building history.")

print(f"\n  Effective annual return for projection: {effective_annual_return:.1f}%")
print("=" * 60)
```

## Total Net Worth Projection

Projects **total net worth** (brokerage + retirement + cash - debts) over the projection period.
- Brokerage grows via compound returns + monthly contributions (with annual raises)
- Retirement accounts grow at their own assumed rate + any monthly contributions
- Cash stays flat (emergency fund maintained, not invested)

```python
# Month-by-month projection — FULL NET WORTH
monthly_rate = effective_annual_return / 100 / 12
retirement_monthly_rate = RETIREMENT_ANNUAL_RETURN / 100 / 12
projection_months = PROJECTION_YEARS * 12

rows = []
brokerage_bal = current_value
retirement_bal = float(RETIREMENT_HSA)
cash_bal = float(CASH_SAVINGS)
debt_bal = float(DEBTS)
cumulative_brokerage_contrib = 0.0
cumulative_retirement_contrib = 0.0
starting_brokerage = current_value
monthly_brokerage_contrib = float(MONTHLY_CONTRIBUTION)
monthly_retirement_contrib = float(RETIREMENT_MONTHLY_CONTRIBUTION)
current_year = datetime.now().year

# Record month 0 (today)
rows.append({
    "month": 0,
    "year": current_year,
    "age": CURRENT_AGE,
    "brokerage": brokerage_bal,
    "retirement": retirement_bal,
    "cash": cash_bal,
    "debts": debt_bal,
    "net_worth": brokerage_bal + retirement_bal + cash_bal - debt_bal,
    "cumulative_brokerage_contrib": 0.0,
    "cumulative_retirement_contrib": 0.0,
    "brokerage_growth": 0.0,
    "retirement_growth": 0.0,
})

for m in range(1, projection_months + 1):
    # Increase contributions at the start of each new year
    if m > 1 and m % 12 == 1:
        monthly_brokerage_contrib *= (1 + ANNUAL_CONTRIBUTION_INCREASE / 100)
        monthly_retirement_contrib *= (1 + ANNUAL_CONTRIBUTION_INCREASE / 100)

    # Brokerage: compound growth + contribution
    brokerage_bal = brokerage_bal * (1 + monthly_rate) + monthly_brokerage_contrib
    cumulative_brokerage_contrib += monthly_brokerage_contrib

    # Retirement/HSA: compound growth + contribution
    retirement_bal = retirement_bal * (1 + retirement_monthly_rate) + monthly_retirement_contrib
    cumulative_retirement_contrib += monthly_retirement_contrib

    # Cash stays flat; debts stay flat (pay off separately)
    net_worth = brokerage_bal + retirement_bal + cash_bal - debt_bal

    rows.append({
        "month": m,
        "year": current_year + ((m - 1) // 12),
        "age": CURRENT_AGE + m / 12,
        "brokerage": brokerage_bal,
        "retirement": retirement_bal,
        "cash": cash_bal,
        "debts": debt_bal,
        "net_worth": net_worth,
        "cumulative_brokerage_contrib": cumulative_brokerage_contrib,
        "cumulative_retirement_contrib": cumulative_retirement_contrib,
        "brokerage_growth": brokerage_bal - starting_brokerage - cumulative_brokerage_contrib,
        "retirement_growth": retirement_bal - RETIREMENT_HSA - cumulative_retirement_contrib,
    })

proj_df = pd.DataFrame(rows)

final = proj_df.iloc[-1]
print(f"Projected net worth at age {TARGET_FI_AGE}: ${final['net_worth']:,.0f}")
print(f"")
print(f"  Brokerage portfolio:      ${final['brokerage']:,.0f}")
print(f"    Starting:                 ${starting_brokerage:,.0f}")
print(f"    Contributions:            ${final['cumulative_brokerage_contrib']:,.0f}")
print(f"    Investment growth:        ${final['brokerage_growth']:,.0f}")
print(f"  Retirement / HSA:         ${final['retirement']:,.0f}")
print(f"    Starting:                 ${RETIREMENT_HSA:,.0f}")
print(f"    Contributions:            ${final['cumulative_retirement_contrib']:,.0f}")
print(f"    Growth:                   ${final['retirement_growth']:,.0f}")
print(f"  Cash / savings:           ${final['cash']:,.0f}")
if DEBTS > 0:
    print(f"  Credit card debt:        -${final['debts']:,.0f}")
print(f"  {'─' * 40}")
print(f"  TOTAL NET WORTH:          ${final['net_worth']:,.0f}")
```

```python
# Net worth projection chart — stacked by account type
fig, ax = plt.subplots(figsize=(14, 7))

months = proj_df["month"]

# Stacked areas: cash (bottom) → retirement → brokerage (top)
# Debts shown as offset from cash layer
net_cash = proj_df["cash"] - proj_df["debts"]
ax.fill_between(months, 0, net_cash,
                color="#64748b", alpha=0.5, label=f"Cash net of debt (${CASH_SAVINGS - DEBTS:,.0f})")
ax.fill_between(months, net_cash,
                net_cash + proj_df["retirement"],
                color="#8b5cf6", alpha=0.5, label="Retirement / HSA")
ax.fill_between(months, net_cash + proj_df["retirement"],
                proj_df["net_worth"],
                color="#22c55e", alpha=0.5, label="Brokerage Portfolio")

# Net worth line on top
ax.plot(months, proj_df["net_worth"], color="#fff", linewidth=2.5, alpha=0.9, label="Total Net Worth")

# Milestone lines
milestones = [250_000, 500_000, 750_000, 1_000_000]
milestone_colors = ["#22c55e", "#eab308", "#f97316", "#ef4444"]
max_val = proj_df["net_worth"].max()
for target, color in zip(milestones, milestone_colors):
    if target <= max_val * 1.1:
        ax.axhline(y=target, color=color, linestyle="--", alpha=0.4, linewidth=1)
        ax.text(projection_months * 0.02, target * 1.02,
                f"${target / 1000:.0f}K" if target < 1_000_000 else "$1M",
                color=color, fontsize=10, alpha=0.7)

# Annotate final net worth
ax.annotate(
    f"${final['net_worth']:,.0f}\nAge {TARGET_FI_AGE}",
    xy=(projection_months, final["net_worth"]),
    xytext=(-100, 15), textcoords="offset points",
    fontsize=12, fontweight="bold", color="#fff",
    arrowprops=dict(arrowstyle="->", color="#fff"),
)

# Formatting
ax.set_title(f"Total Net Worth Projection: Age {CURRENT_AGE} to {TARGET_FI_AGE}",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Years from Now")
ax.set_ylabel("Net Worth ($)")
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))

# X axis: show years + age
year_ticks = list(range(0, projection_months + 1, 12))
year_labels = [f"Yr {y // 12}\n(Age {CURRENT_AGE + y // 12})" for y in year_ticks]
ax.set_xticks(year_ticks)
ax.set_xticklabels(year_labels, fontsize=8)

ax.legend(loc="upper left", fontsize=10)
ax.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```

## Year-by-Year Breakdown

```python
# Extract year-end snapshots
yearly = proj_df[proj_df["month"] % 12 == 0].copy()
if proj_df.iloc[-1]["month"] % 12 != 0:
    yearly = pd.concat([yearly, proj_df.iloc[[-1]]])

print("=" * 105)
print(f"{'Year':>6} {'Age':>5} {'Brokerage':>14} {'Retirement':>14} {'Cash':>10} {'Net Worth':>14} {'YoY Change':>14}")
print("=" * 105)

prev_nw = total_net_worth

for _, row in yearly.iterrows():
    if row["month"] == 0:
        yoy = ""
        print(f"{int(row['year']):>6} {CURRENT_AGE:>5} ${row['brokerage']:>12,.0f} ${row['retirement']:>12,.0f} ${row['cash']:>8,.0f} ${row['net_worth']:>12,.0f} {'(today)':>14}")
        prev_nw = row["net_worth"]
        continue

    age = int(CURRENT_AGE + row["month"] / 12)
    year = int(row["year"])
    change = row["net_worth"] - prev_nw
    change_pct = (change / prev_nw * 100) if prev_nw > 0 else 0

    print(f"{year:>6} {age:>5} ${row['brokerage']:>12,.0f} ${row['retirement']:>12,.0f} ${row['cash']:>8,.0f} ${row['net_worth']:>12,.0f} {change_pct:>+12.1f}%")

    prev_nw = row["net_worth"]

print("=" * 105)

total_contrib = final["cumulative_brokerage_contrib"] + final["cumulative_retirement_contrib"]
total_growth = final["brokerage_growth"] + final["retirement_growth"]
print(f"\n  Starting net worth:     ${total_net_worth:,.0f}")
print(f"  Total contributions:    ${total_contrib:,.0f}")
print(f"  Total investment growth:${total_growth:,.0f}")
print(f"  Final net worth:        ${final['net_worth']:,.0f}")
```

## Milestone Tracker

```python
print("=" * 60)
print("         NET WORTH MILESTONE TRACKER")
print("=" * 60)

milestones = [250_000, 500_000, 750_000, 1_000_000, 1_500_000, 2_000_000]

for target in milestones:
    target_label = f"${target:,.0f}"

    if total_net_worth >= target:
        print(f"  {target_label:>12}: Already reached! Current net worth: ${total_net_worth:,.0f}")
        continue

    hit = proj_df[proj_df["net_worth"] >= target]
    if not hit.empty:
        first_hit = hit.iloc[0]
        months_away = int(first_hit["month"])
        age_at = CURRENT_AGE + months_away / 12
        year_at = datetime.now().year + months_away // 12
        print(f"  {target_label:>12}: Age {age_at:.0f} (month {months_away}, ~{year_at})")
    else:
        print(f"  {target_label:>12}: Not reached within {PROJECTION_YEARS}-year projection")

print("=" * 60)
```

### Assumptions & Disclaimer

**Assumptions used in this projection:**
- Brokerage return: applied as constant monthly rate (real markets are volatile)
- Retirement return: separate assumed rate, compounded monthly
- Monthly contributions increase by a fixed percentage annually (income growth)
- Cash/savings held flat as emergency reserve
- No taxes, withdrawals, or major life events factored in
- As real brokerage history builds up (3+ months), the assumed return will be replaced by your actual CAGR
- Retirement and cash balances are manual entries — update them periodically in the profile section

**Re-run this notebook periodically** to update the projection with real growth data.

**This is not financial advice.** Projections are illustrative. Past performance does not guarantee future results.
